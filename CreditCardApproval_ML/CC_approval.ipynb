```python
# Credit Card Approval
import pandas as pd
cc_apps = pd.read_csv('/content/crx.csv',header=None)
cc_apps.tail()

# actual (['b',30.83,0,'u','g','w','v',1.25,'t','t',1,'f','g'])
# modified as 11 and 13 are dropped (['b',30.83,0,'u','g','w','v',1.25,'t','t',1,'g',0])
```

```python
cc_apps_description = cc_apps.describe()
print(cc_apps_description)
print("\n")

cc_apps_info = cc_apps.info()
# print(cc_apps_info)
# print("\n")

cc_apps.tail(17)
```

```python
import numpy as np
print(cc_apps.isnull().values.sum())
cc_apps = cc_apps.replace('?', np.nan)
print("Total nan: " +str(cc_apps.isnull().values.sum()))
print("Nan by column: " "\n")
print(cc_apps.isnull().sum())
cc_apps.tail(17)
```

```python
cc_apps.mean(numeric_only=True)
cc_apps.fillna(cc_apps.mean(), inplace=True)
print("Total nan: " +str(cc_apps.isnull().values.sum()))
cc_apps.isnull().sum()
```

```python
for col in cc_apps:
    if cc_apps[col].dtypes == 'object':
        cc_apps= cc_apps.fillna(cc_apps[col].value_counts().index[0])
print('total missing values:' +str(cc_apps.isnull().values.sum()))
print('missing values in each column:')
cc_apps.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l= LabelEncoder()
for col in cc_apps:
    if cc_apps[col].dtypes == 'object':
        l.fit(cc_apps[col])
        cc_apps[col]=l.transform(cc_apps[col])
cc_apps.info()
```

```python
from sklearn.model_selection import train_test_split
cc_apps= cc_apps.drop([11,13],axis=1)
cc_apps_values = cc_apps.values
X,y = cc_apps_values[:,0:13], cc_apps_values[:,13]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,random_state=42)
cc_apps.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
rescaledx_train = scaler.fit_transform(X_train)
rescaledx_test = scaler.fit_transform(X_test)
```

```python
#import logisticRegression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(rescaledx_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = logreg.predict(rescaledx_test)
print("accuracy of the logistic regression classifier: ", logreg.score(rescaledx_test, y_test))
print("confution matrix : \n", confusion_matrix(y_test,y_pred))
```

```python
import numpy as np

input_data=np.array(['b',30.83,0,'u','g','w','v',1.25,'t','t',1,'g',0])
input_data = l.fit_transform(input_data)
print(input_data)
input_data_reshaped = input_data.reshape(1,-1)
prediction = logreg.predict(input_data_reshaped)

print(prediction)

if(prediction[0]==0.0):
    print('Credit Card accepted')
else:
    print('Credit Card is not accepted')
```
